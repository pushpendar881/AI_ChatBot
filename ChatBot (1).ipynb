```python
from transformers import pipeline
import requests
from PIL import Image
from io import BytesIO

# Initialize the image-to-text pipeline with the specified model
image_pipeline = pipeline("image-to-text", model="Salesforce/blip-image-captioning-large")

# Example function to invoke the text generation model
def invoke_text_generation(prompt):

    from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model directly

    repo_id="mistralai/Mistral-7B-Instruct-v0.3"
    llm=HuggingFaceEndpoint(repo_id=repo_id , task="text-generation",max_length=200,temperature=0.7,token=sec_key)
    response = llm.invoke(prompt)
    return response

# ... other code ...

def handle_input(user_input):
    # Check if the input is an image path or text
    if user_input.startswith("/content/"):  # Example: assuming image path starts with "/content/"
        image_path = user_input
        # Load the image
        try:
            image = Image.open(image_path)
        except Exception as e:
            return f"Error loading image: {e}"

        # Generate captions for the image
        captions = image_pipeline(image_path)
        if not captions:
            return "Could not generate captions for the image."

        # Access the 'generated_text' key instead of 'caption'
        image_context = captions[0]['generated_text']

        # Ask for a question related to the image
        user_question = input("Ask your question related to the image: ")
        combined_prompt = f"Based on the image: {image_context}. Answer the following question: {user_question}"
        response = invoke_text_generation(combined_prompt)
    else:
        # Directly handle the text input
        response = invoke_text_generation(user_input)

    return response

# ... other code ...

# Main function to run the chatbot
def chatbot():
    print("Chatbot: Hello! How can I assist you today?")
    while True:

        print("\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\")
        user_input = input("User: ")
        response = handle_input(user_input)

        print("Chatbot:", response)
        print("\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\")
        if(user_input=="bye"):
            break


# Example usage of the chatbot function
if __name__ == "__main__":
    chatbot()
```
